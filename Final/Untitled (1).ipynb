```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


data = pd.read_csv('Final.csv')


province_counts = data['Province'].value_counts()

# Select the top 5 provinces
top_provinces = province_counts.head(5)

# bar plot for the top 5 provinces
plt.figure(figsize=(10, 6))
sns.barplot(x=top_provinces.index, y=top_provinces.values)
plt.title('Top 5 Provinces by Number of Records')
plt.xlabel('Province')
plt.ylabel('Number of Records')
plt.xticks(rotation=45)

# Annotate each bar with its value
for index, value in enumerate(top_provinces):
    plt.text(index, value + 1, str(value), ha='center', va='bottom')

plt.show()
```
